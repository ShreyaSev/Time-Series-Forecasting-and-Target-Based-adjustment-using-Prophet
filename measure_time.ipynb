```python
start_time = time.time()
for row in df.index:
    series = df.iloc[row]
    temp_df = pd.DataFrame({'ds':series.index, 'y':series.values})
    item = temp_df.iloc[2].values[1]
    
    temp_df = temp_df.drop(0, axis = 0)
    temp_df = temp_df.drop(1, axis = 0)
    temp_df = temp_df.drop(2, axis = 0)
    temp_df = temp_df.sort_values(by = 'ds')

    m = Prophet()
    m.fit(temp_df)
    future = m.make_future_dataframe(periods = 12, freq='MS')
    fcst = m.predict(future)
    outputs_df[item] = fcst['yhat']
end_time = time.time()
```
