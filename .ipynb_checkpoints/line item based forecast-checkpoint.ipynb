```python
bank_df = df[df['Display Caption']=='Bank']
bank_df = bank_df.drop(['Display Caption'], axis = 1)
```

```python
bank_df
```

```python
bank_df = bank_df.T
bank_df.head()
```

```python
bank_df.columns = bank_df.iloc[1]
bank_df = bank_df.iloc[2:]
```

```python
bank_df
```

```python
outputs_df = pd.DataFrame()
```

```python
for line_item in bank_df.columns:
    m = Prophet()
    item_df = bank_df[line_item]
    item_df = pd.DataFrame(item_df)
    item_df.reset_index(inplace=True)
    item_df.columns = ['ds', 'y']
    m.fit(item_df)
    future = m.make_future_dataframe(periods = 12, freq = 'MS')
    fcst = m.predict(future)
    outputs_df[line_item] = fcst['yhat']
```

```python
outputs_df.head()
```

```python
future.tail()
```

```python
outputs_df.index = future['ds']
outputs_df
```

```python
month_wise = outputs_df.T
```

```python
monthwise_sum = month_wise.sum()
cur_sum = monthwise_sum.sum()
target = cur_sum*10
already_achieved = cur_sum - monthwise_sum[-12:].sum()
cur_future_sum = cur_sum-already_achieved
```

```python
cur_future_sum
```

```python
rem_target = target - already_achieved
rem_target
```

```python
new_monthwise_sum = []

for val in monthwise_sum[-12:]:
    new_val = (val/cur_future_sum * rem_target)
    new_monthwise_sum.append(new_val)
```

```python
def adjust_data(temp_df, month, caption_target):
    caption_sum = temp_df.sum().values[0]
    #caption_target = eval(input('Target for given month: '))
    new_fcst = pd.DataFrame(columns = [month])
    for i in temp_df.index:
        new_fcst.loc[i] =(temp_df.loc[i].values[0]/caption_sum) * caption_target
    return new_fcst
```

```python
new_fcst = adjust_data(temp_df)
```

```python
def adjust_month(month_wise, month, month_target):
    temp_df = month_wise.loc[:,month]
    temp_df = pd.DataFrame(temp_df)
    new_fcst = adjust_data(temp_df, month, month_target)
    return new_fcst
```

```python
new_fcst = adjust_month(month_wise, '2020-03-01')
```
